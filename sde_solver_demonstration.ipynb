```python
%matplotlib inline
```

```python
import optoanalysis.sde_solver as sde_solver
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Gamma0 = 4000 # radians/second
Omega0 = 75e3*2*np.pi # radians/second
eta = 0.0
T_0 = 300 # K
m = 3.1e-19 # KG
```

```python
sim1 = sde_solver.sde_solver(Omega0, Gamma0, m, seed=88, TimeTuple=[0, 0.01], dt=1e-9)
```

```python
sim1.solve()
```

```python
%timeit sim1.solve()
```

```python
plt.plot(sim1.tArray*1e6, sim1.q)
plt.xlabel("t (us)")
plt.ylabel("q")
```

```python
plt.plot(sim1.tArray*1e6, sim1.v)
plt.xlabel("t (us)")
plt.ylabel("v")
```

```python
import optoanalysis
```

```python
f, psd = optoanalysis.calc_PSD(sim1.q, 1/sim1.dt)
```

```python
plt.semilogy(f/1e3, psd)
plt.xlim([0, 500])
```
